```python
f_new2 = data.copy()
f_new2 = pd.get_dummies(data)
f_new2 =  f_new2.drop("Gender_M",1)
```

```python
from sklearn.preprocessing import normalize
x_norm = normalize(f_new2)
x_norm
```

```python
Km.fit(x_norm)
print(Km.labels_)
    
```

```python
a = []
for i in range(2,10):
    km_new = KMeans(n_clusters=i,max_iter=10 )
    km_new.fit(f_new2)
    a.append(km_new.inertia_)
    
plt.plot(range(2,10),a,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
```

```python
# So the no of clusters as per the method is 6
```

# Silhoutte Score 

```python
from sklearn.metrics import silhouette_score

for i in range(2,15):
    kmeans = KMeans(n_clusters=i, max_iter=50)
    kmeans.fit(x_norm)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(x_norm, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(i, silhouette_avg))
```

```python
# As per Silhoutte score the value is 6 
```

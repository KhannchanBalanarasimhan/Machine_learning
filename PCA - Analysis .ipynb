```python
from sklearn.ensemble import RandomForestRegressor


model_rf=RandomForestRegressor(n_estimators=150)


model_rf.fit(x_pca,y_train)
```

```python
train_pred=model_rf.predict(x_pca)
train_pred
```

```python
test_pred = model_rf.predict(x_test_pca)
test_pred
```

```python
model_rf.score(x_test_pca,y_test)
```

```python
# PCA from Scratch
```

```python
features = mydata.T
cov_matrix = np.cov(features)
cov_matrix[:5]
```

```python
values, vectors = np.linalg.eig(cov_matrix)
values[:5]
```

```python
vectors[:5]
```

```python
explained_variances = []
for i in range(len(values)):
    explained_variances.append(values[i] / np.sum(values))
 
print(np.sum(explained_variances),"\n", explained_variances)
```

```python
# AS we can see a lot of data has caused the overfitting problem
```

```python
projected_1 = mydata.dot(vectors.T[0])
projected_2 = mydata.dot(vectors.T[1])
res = pd.DataFrame(projected_1, columns=['PC1'])
res["PC2"] = projected_2
res['SalesPricee for the house'] = y_dep
res.head()
```

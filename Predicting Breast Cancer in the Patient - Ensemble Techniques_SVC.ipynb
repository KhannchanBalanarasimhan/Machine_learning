```python
# OUTLIER REMOVAL
```

```python
df = data.drop("diagnosis",1)
```

```python
for i in df.columns:
    sns.boxplot(df[i])
    plt.show()
```

```python
# Removal of outliers 
```

```python
for values in df.columns:
    
    q1 = np.percentile( df[values], 25, interpolation = 'midpoint')
    q3 = np.percentile( df[values], 75, interpolation = 'midpoint')
    
    iqr= q3-q1
    
    df[values].drop(df[(df[values]>=(q3+1.5*iqr))|(df[values]<=(q1-1.5*iqr))].index,inplace=True)
```

```python
for i in df.columns:
    sns.kdeplot(df[i])
    plt.show()
```
